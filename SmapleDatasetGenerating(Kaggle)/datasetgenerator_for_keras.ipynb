### This generate sample datasets form main kaggle dataset for CNN keras.

```python

import cv2
from matplotlib import pyplot as plt
import numpy as np
 
import pandas as pd
```

```python
df= pd.read_csv("trainLabels.csv")
df
```

```python
size=600
df0=  df[df['level'] ==0].head(size) 
df1=  df[df['level'] ==1].head(size) 
df2=  df[df['level'] ==2].head(size) 
df3=  df[df['level'] ==3].head(size) 
df4=  df[df['level'] ==4].head(size) 
df4
```

There is a folder called 'train' which contains all images of dataset

```python
def generate(data,level):
    for name in data.image:
        image = cv2.imread('./train/' + name + '.jpeg', cv2.IMREAD_UNCHANGED)
        #crop image (remove black space)     
        grayscale = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        ret,thresholded = cv2.threshold(grayscale, 0, 255, cv2.THRESH_OTSU)
        bbox = cv2.boundingRect(thresholded)
        x, y, w, h = bbox
        foreground = image[y:y+h, x:x+w]

        
        resized = cv2.resize(foreground, (300,300),interpolation = cv2.INTER_AREA)
        # CLAHE on all channels
        (B, G, R) = cv2.split(resized)

        clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))

        c0 = clahe.apply(B)
        c1 = clahe.apply(G)
        c2 = clahe.apply(R)

        merged = cv2.merge([c0, c1, c2])
        
        cv2.imwrite('./newdataset/'+ level +'/'+name + '.jpeg', merged )
    
```

```python
generate(df0,'0')
generate(df1,'1')
generate(df2,'2')
generate(df3,'3')
generate(df4,'4')
```
