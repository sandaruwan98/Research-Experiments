```python

import cv2
from matplotlib import pyplot as plt
import numpy as np
 
import pandas as pd
```

```python
def generate(data,level,folder):
    for name in data['Image name']:
        image = cv2.imread(folder+ '/' + name, cv2.IMREAD_UNCHANGED)
#         print(name)
        #crop image (remove black space)     
        grayscale = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        ret,thresholded = cv2.threshold(grayscale, 0, 255, cv2.THRESH_OTSU)
        bbox = cv2.boundingRect(thresholded)
        x, y, w, h = bbox
        foreground = image[y:y+h, x:x+w]

        
        resized = cv2.resize(foreground, (512,512),interpolation = cv2.INTER_AREA)
        outfile = name.split('.')[0] + '.jpg'
#         print(outfile)
        cv2.imwrite('./withoutclahe/'+ level +'/'+outfile, resized )
    
```

```python


def generatefolder(foldername):

    df= pd.read_excel(foldername + "/Annotation.xls")

    size=20
    df0=  df[df['Retinopathy grade'] ==0].head(size) 
    df1=  df[df['Retinopathy grade'] ==1].head(size) 
    df2=  df[df['Retinopathy grade'] ==2].head(size) 
    df3=  df[df['Retinopathy grade'] ==3].head(size) 
    

    generate(df3,'3',foldername)
```
